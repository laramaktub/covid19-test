***IMÁGENES SIN ETIQUETA***


```python
import random

sin_etiqueta=list(range(94))
random.seed(30)
random_sin_etiqueta=random.sample(sin_etiqueta,94)
res = list(enumerate(random_sin_etiqueta))
print(res)
```

***IMÁGENES CON ETIQUETA***

```python
con_etiqueta=list(range(94))
random.seed(42)
random_con_etiqueta=random.sample(con_etiqueta,94)
res = list(enumerate(random_con_etiqueta))
print(res)
```
