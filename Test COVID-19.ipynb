***IMÁGENES SIN ETIQUETA***


```python
import random
from csv import reader


original=list(range(94))

original_labels=[]
with open('Labels.csv', 'r') as read_obj:
    csv_reader = reader(read_obj)
    header = next(csv_reader)
    # Check file as empty
    i=0
    if header != None:
        for row in csv_reader:
            label="Covid19"
            if row[2]=='0':
                label="Normal"
            original_labels.append(label)
            i+=1
            

sin_etiqueta = list(zip(original,original_labels))

random.seed(30)
random_sin_etiqueta=random.sample(original,94)
j=0
for idx, val in enumerate(random_sin_etiqueta):
    print("Image_"+str(idx), sin_etiqueta[val])
    j+=1

```
